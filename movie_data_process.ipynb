```python
import pandas as pd
import numpy as npc

movies_df = pd.read_csv("data/tmdb_5000_movies.csv")
# drop columns
# budget, homepage, original_language, original_title, production_countries, release_date,
# revenue, spoken_languages, status
movies_df = movies_df.drop(columns=['budget', 'homepage', 'original_language', 'original_title', 'production_countries',
                                    'release_date', 'revenue', 'spoken_languages', 'status'])
```

```python
credits_df = pd.read_csv("data/tmdb_5000_credits.csv")
credits_df = credits_df.drop(columns=['title'])
credits_df.columns=['id', 'cast', 'crew']
credits_df.head()
```

```python
total_df = movies_df.join(credits_df.set_index('id'), on='id')
```

```python
#total_df.to_csv("data/tmdb_5000_totals.csv")
total_df['cast'][:100]
```

```python
total_df.info()
```

```python
import json

json_columns = ['genres', 'keywords', 'cast', 'crew', 'production_companies']
for column in json_columns:
    total_df[column] = total_df[column].apply(json.loads)
```

```python
# there are 43 empty cast - remove those rows
count = 0
for title, crew_list in zip(total_df['title'], total_df["crew"]):
    if not crew_list:
        print(title,'       ',crew_list)
        total_df = total_df[total_df.title != title]
        count+=1
print(count)
```

```python
count=0
for title, cast_list in zip(total_df['title'], total_df["crew"]):
    if not cast_list:
        print(title)
        count+=1
print(count)
```

```python
director_list = list()

for title, crew_list in zip(total_df['title'], total_df['crew']):
    temp = list()
    for one_crew_dict in crew_list:
        if one_crew_dict['job'] == 'Director':
            temp.append(one_crew_dict['name'])
    if not temp:
        total_df = total_df[total_df.title != title]
        break
    director_list.append(temp)

directors = list()
for item in director_list:
    if len(item) >= 1:
        directors.append(item[0])
    else:
        print("No director")
total_df['director'] = directors
```

```python
# test director process successful
count = 0
for director_name, crew_list in zip(total_df['director'], total_df['crew']):
    for one_crew in crew_list:
        if one_crew['job'] == 'Director':
            if one_crew['name'] == director_name:
                count+=1
                break
print(count)
```

```python
# drop crew column from total_df
total_df = total_df.drop(columns=['crew'])
```
